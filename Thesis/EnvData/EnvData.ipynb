```python
import pandas as pd
```

```python
env = pd.read_excel(r'C:\Users\ER2C-02\Google Drive\NoteBook\Thesis\EnvData\GHdata.xlsx')
```

```python
env.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(11,5))
x = env.No

plt.title("Temperature")
plt.xlabel("Time(m)")
plt.ylabel("Degree (C)")

plt.plot(x,env.OutTemp, label='Air Temperature')
plt.plot(x,env.soilTemp, label='Soil temperature')

plt.legend()

plt.show()
#plt.savefig("foo.png")
```

```python
fig = plt.figure(figsize=(11,5))
x = env.No

plt.title("Relative Humidity")
plt.xlabel("Time(m)")
plt.ylabel("Relative Humidity (%)")

plt.plot(x,env.OutHum)

#plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(11,5))
x = env.No

plt.title("Temperature, Relative humidity vs time")
plt.xlabel("Time(m)")
plt.ylabel("Temp(C) & RH (%)")

plt.plot(x,env.OutTemp, label='Temperature')
plt.plot(x,env.OutHum, label='Humidity')

plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(11,5))
x = env.No
plt.title("Soil VWC")
plt.xlabel("Time(m)")
plt.ylabel("%")
plt.plot(x,env.soil_vwc, label='Soil water content')
#plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(11,5))
x = env.No
plt.title("Soil Electro Conductivity")
plt.xlabel("Time(m)")
plt.ylabel("EC (S/m)")
plt.plot(x,env.soil_ec, label='Temperature')
plt.show()
```

```python
fig = plt.figure(figsize=(11,5))
x = env.No
plt.title("Soil VWC")
plt.xlabel("Time(m)")
plt.ylabel("EC (S/m)")
plt.plot(x,env.soil_salinity, label='Temperature')
plt.show()
```
