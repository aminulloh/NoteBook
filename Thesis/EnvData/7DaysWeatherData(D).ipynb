# Weather Node data in 7 Days

```python
import pandas as pd
import sqlite3
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,5))
```

```python
df = pd.read_csv("Weather_data.csv", parse_dates=["timestamp"], index_col="timestamp")
df.head(5)
```

## Data characteristics in a day 

### Temperature data in (2019-11-20)

```python
fig = plt.figure(figsize=(10,5))
plt.title("Temperature")
plt.xlabel("Time")
plt.ylabel("Degree (C)")
df["2019-11-20"].tempSHT.plot(label='Sensor SHT')
df["2019-11-20"].tempMPL.plot(label='Sensor MPL')
plt.legend()
```

### Relative-Humidity data in (2019-11-20)

```python
fig = plt.figure(figsize=(10,5))
plt.title("Relative Humidity")
plt.xlabel("Time")
plt.ylabel("RH (%)")
df["2019-11-20"].hum.plot()
#plt.legend()
```

### Light intensity data in (2019-11-20)

```python
fig = plt.figure(figsize=(10,5))
plt.title("Temperature")
plt.xlabel("Time")
plt.ylabel("Degree (C)")
df["2019-11-20"].IR.plot(label='IR')
df["2019-11-20"].Full.plot(label='Full')
df["2019-11-20"].Visible.plot(label='Visible')
df["2019-11-20"].Lux.plot(label='Lux')
plt.legend()
```

### Air data in (2019-11-20)

```python
fig = plt.figure(figsize=(10,5))
plt.title("Equal Carbondioxide eCO2")
plt.xlabel("Time")
plt.ylabel("ppm")
df["2019-11-20"].eCO2.plot()
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Equal Carbondioxide eCO2")
plt.xlabel("Time")
plt.ylabel("ppm")
df["2019-11-20"].TVOC.plot()
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Barometric Pressure")
plt.xlabel("Time")
plt.ylabel("Hg")
df["2019-11-20"].Pressure.plot()
```

## Weather data in multiple days
