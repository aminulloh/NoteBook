# Tomato Growth Measurement

```r
data <-data.frame(day=c(0,6,13,18,26,46,73,94,105,125),
                  leaf=c(0,2,6,8,17,89,336,696,761,801),
                  WiltLeaf=c(NA,NA,NA,NA,NA,NA,0,46,124,156),
                  nStem=c(0,0,2,3,5,13,40,80,112,149),
                  nStemFall=c(NA,NA,NA,NA,NA,NA,0,4,6,9),
                  PlantHeight=c(0,3,5,7,13,53,132,147,167,190),
                  BlossomInit=c(NA,NA,NA,NA,NA,0,42,157,207,229),
                  BlossomDrop=c(NA,NA,NA,NA,NA,0,27,113,193,212),
                  FruitHeight=c(NA,NA,NA,NA,NA,NA,NA,0,13.8,19.3),
                  FruitDia=c(NA,NA,NA,NA,NA,NA,NA,0,12,22.5))
```

```r
# Leaves
plot(data$day,data$leaf, xlab="Day",ylab=" Numbers (n)",main="Leaves initiation",type="b",col = "red", lty=1,lwd=3)
lines(data$day,data$WiltLeaf, type="o",col = "blue",lty=2,lwd=3)
legend("topleft",c("Leaves initiated","Wilt leaves"),lty=c(1,2),col=c("red","blue"),lwd=c(3,3))
```

```r
# Stems
plot(data$day,data$nStem, xlab="Day",ylab=" Numbers (n)",main="Stems initiation",type="b",col = "red",lty=1,lwd=3)
lines(data$day,data$nStemFall, type="o",col = "blue",lty=2,lwd=3)
legend("topleft",c("Stems initiated","Stem fall"),lty=c(1,2),col=c("red","blue"),lwd=c(3,3))
```

```r
# Blossom
plot(data$day,data$BlossomInit, xlab="day",ylab=" Milimeters (mm)",main="Blossoms initiation ",type="b",col = "blue",lty=1,lwd=3)
lines(data$day,data$BlossomDrop ,type="o",col = "red",lty=2,lwd=3)
legend("topleft",c("Blossoms initiation","Blossom fall"),col=c("blue","red"),lty=c(1,2),lwd=c(3,3))
```

```r
# Blossom
plot(data$day,data$BlossomInit, xlab="day",ylab=" Milimeters (mm)",main="Blossoms initiation ",type="b",col = "blue",lty=1,lwd=3)
lines(data$day,data$BlossomDrop ,type="o",col = "red",lty=2,lwd=3)
legend("topleft",c("Blossoms initiation","Blossom fall"),col=c("blue","red"),lty=c(1,2),lwd=c(3,3))
```

```r
# Fruit
plot(data$day,data$FruitDia, xlab="day",ylab=" Milimeters (mm)", main="Fruit growth", col = "red", type="b",lty=1,lwd=3)
lines(data$day,data$FruitHeight,type="o",col = "blue",lty=2,lwd=3)
legend("topleft",c("Fruit Diameter","Fruit Height"),col=c("red","blue"),lty=c(1,2),lwd=c(3,3))
```

```r
# Growth Model
duration=125
x=seq(25,duration,by=5)
```

```r
# Leaves growth model 
LL=800
kL=0.085
zL = 73
f_Leaves = LL / (1 + exp(-kL * (x-zL)))
plot(data$day,data$leaf, type="b",col = "blue", xlab="Day",ylab=" Numbers (n)",main="Leaves initiation trajectory",lty=3,lwd=3)
lines(x,f_Leaves, type="l",col = "red", lty=1,lwd=3)
legend("topleft",c("Measurement","Growth model"),lty=c(3,1),col=c("blue", "red"),lwd=c(3,3))
```

```r
# Stems growth model 
LS=175
kS=0.05
zS = 95
f_Stem = LS / (1 + exp(-kS * (x-zS)))
plot(data$day,data$nStem, type="b",col = "blue", xlab="Day",  ylab=" Numbers (n)",main="Stems initiation trajectory",lty=3,lwd=3)
lines(x,f_Stem, type="l",col = "red", lty=1,lwd=3)
legend("topleft",c("Stem measurement","Stem growth model"),lty=c(3,1),col=c("blue", "red"),lwd=c(3,3))
```

```r
# Plant height growth model 
LH=200
kH=0.06
zH = 65
f_Height = LH / (1 + exp(-kH * (x-zH)))
plot(data$day,data$PlantHeight, type="b",col = "blue", xlab="Day",  ylab="Height (cm)",main="Plant height trajectory",lty=3,lwd=3)
lines(x,f_Height, type="l",col = "red", lty=1,lwd=3)
legend("topleft",c("measurement","growth model"),lty=c(3,1),col=c("blue", "red"),lwd=c(3,3))
```

```r
# Blossoms growth model 
LB=235
kB=0.1
zB = 85
f_Blos = LB / (1 + exp(-kB * (x-zB)))
plot(data$day,data$BlossomInit, type="b",col = "blue", xlab="Day",  ylab=" Numbers (n)",main="Blossoms initiation trajectory",lty=3,lwd=3)
lines(x,f_Blos, type="l",col = "red", lty=1,lwd=3)
legend("topleft",c("Measurement","Growth model"),lty=c(3,1),col=c("blue", "red"),lwd=c(3,3))
```

```r
# Fruit growth model 
LF=23
kF=0.16
zF = 105
f_Fruit = LF / (1 + exp(-kF * (x-zF)))
plot(data$day,data$FruitDia, type="b",col = "blue", xlab="Day",  ylab=" Diameters (cm)",main="Fruit Development",lty=3,lwd=3)
lines(x,f_Fruit, type="l",col = "red", lty=1,lwd=3)
legend("topleft",c("Measurement","Growth model"),lty=c(3,1),col=c("blue", "red"),lwd=c(3,3))
```

```r
plot(data$day,data$FruitDia,type="b", lwd=2, xlab="Days",ylab="Fruit Diameter (mm)",
     main="Fruit Growth in Diameter")
```

```r
#----------------Model Gabung----------------#
# Leaves growth and blossoms model 
LL=800
kL=0.085
zL = 73
f_Leaves = LL / (1 + exp(-kL * (x-zL)))
LB=235
kB=0.1
zB = 85
f_Blos = LB / (1 + exp(-kB * (x-zB)))
LF=23
kF=0.16
zF = 105
f_Fruit = LF / (1 + exp(-kF * (x-zF)))
plot(data$day,data$leaf, type="p",col = "red",pch=3, ylab="", xlab="Day",main="Leaves, blossoms and fruit trajectory",lwd=3)
lines(x,f_Leaves, type="l",col = "red", lty=1,lwd=3)
lines(data$day,data$BlossomInit, type="p",col = "green",pch=5,lwd=3)
lines(x,f_Blos, type="l",col = "green", lty=3,lwd=3)
lines(data$day,data$FruitDia, type="p",col = "blue",pch=2,lwd=3)
lines(x,f_Fruit, type="l",col = "blue", lty=5,lwd=3)
legend("topleft",c("Leaf initiation measurement (n)","Leaf initiation model (n)","Blossoms initiation measurement (n)","Blossoms initiation model (n)",
                   "Fruit diameter measurement (mm)","Fruit diameter model (mm)")
       ,lty=c(0,2,0,3,0,4),pch=c(3,NA,5,NA,2,NA) ,col=c("red", "red","green", "green","blue","blue"),lwd=c(3,3))
```

```r
# Plant height growth model 
LH=200
kH=0.06
zH = 65
f_Height = LH / (1 + exp(-kH * (x-zH)))
LS=175
kS=0.05
zS = 95
f_Stem = LS / (1 + exp(-kS * (x-zS)))
plot(data$day,data$PlantHeight, type="p",pch=3,col = "red", xlab="Day",  ylab="",main="Plant height and stems initiation trajectory",lty=3,lwd=3)
lines(x,f_Height, type="l",col = "red", lty=2,lwd=3)
lines(data$day,data$nStem, type="p",pch=2,col = "blue",lty=3,lwd=3)
lines(x,f_Stem, type="l",col = "blue", lty=3,lwd=3)
legend("topleft",c("Plant heigh measurement (cm)","Plant heigh model (cm)","Stems initiation measurement (n)",
                   "Stems initiation model (n)"),lty=c(NA,2,NA,3),pch=c(3,NA,2,NA),col=c("red", "red","blue", "blue"),lwd=c(3,3))
```

```r
# Stage
plot(x,f_Leaves, xlab="Day",ylab="",main="Tomato plant growth stage", col = "green", type="l",lty=1,lwd=3)
lines(x,f_Height, type="l",col = "black",lty=1,lwd=3)
lines(x,f_Stem, type="l",col = "blue",lty=1,lwd=3)
lines(x,f_Blos, type="l",col = "magenta",lty=1,lwd=3)
lines(x,f_Fruit, type="l",col = "red",lty=1,lwd=3)
legend("topleft",c("Plant height (cm)","Leaves initiation (n)","Stems initiation (n)","Blossoms initiation (n)","Fruit diameter (cm)"),
       lty=c(1),lwd=c(3,3,3,3,3),col=c("black","green","blue","magenta","red"))
```
